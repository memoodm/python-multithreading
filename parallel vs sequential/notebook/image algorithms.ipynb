```python
import imageio.v2 as iio
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
IMAGE = f"C:\\Users\\memoo\\Desktop\\python\\parallel\\img\\pikachu.jpg"
```

```python
img = iio.imread(IMAGE)
plt.imshow(img)
```

```python
def image_to_grayscale(img):
    result_image = np.zeros_like(img)
    for row in range(len(img)):
        for column in range(len(img[row])):
            pixel = img[row][column]
            gray_value = np.mean(pixel)
            result_image[row][column] = [gray_value, gray_value, gray_value]
    return result_image

def image_to_silhouette(img):
    result_image = np.zeros_like(img)
    for row in range(len(img)):
        for column in range(len(img[row])):
            pixel = img[row][column]
            result_image[row][column] = [0, 0, 0] if 100 > np.sum(pixel) else [255, 255, 255]
    return result_image

def image_to_invert(img):
    result_image = np.zeros_like(img)
    for row in range(len(img)):
        for column in range(len(img[row])):
            pixel = img[row][column]
            result_image[row][column] = [ 255-pixel[0], 255-pixel[1], 255-pixel[2] ]
    return result_image    

def execution(img, method):
    inicio = time.perf_counter()
    result_image = method(img)
    fin = time.perf_counter()
    execution_time = fin - inicio
    print(f"Tiempo de ejecución: {fin - inicio:.6f} segundos")
    return result_image, execution_time
```

```python
result_image, execution_time = execution(img, method=image_to_grayscale)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle(f"Tiempo de ejecucion {execution_time:.6f} segundos")
axes[0].imshow(img)
axes[0].set_title("Original")
axes[1].imshow(result_image)
axes[1].set_title("Result")
plt.show()
```

```python
result_image, execution_time = execution(img, method=image_to_silhouette)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle(f"Tiempo de ejecucion {execution_time:.6f} segundos")
axes[0].imshow(img)
axes[0].set_title("Original")
axes[1].imshow(result_image)
axes[1].set_title("Result")
plt.show()
```

```python
result_image, execution_time = execution(img, method=image_to_invert)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
plt.suptitle(f"Tiempo de ejecucion {execution_time:.6f} segundos")
axes[0].imshow(img)
axes[0].set_title("Original")
axes[1].imshow(result_image)
axes[1].set_title("Result")
plt.show()
```
